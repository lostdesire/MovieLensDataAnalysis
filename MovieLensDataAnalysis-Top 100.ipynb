## Make a list of top 100 most popular movies

---

### A_1) highest number of ratings for all users

```python
all_1 = data[["movie id", "rating"]]
all_1 = all_1["rating"].groupby(all_1["movie id"]).count()
all_1 = all_1.sort_values(ascending=False)
```

```python
#all_1
```

```python
all_2 = item[["movie id", "movie title"]]
```

```python
#all_2
```

```python
rank = pd.merge(all_2, all_1, on="movie id", how="left")
rank = rank.sort_values(by="rating", ascending= False).head(100)
rank = rank.reset_index(drop=True)

all_rank = pd.DataFrame({"All users":rank["movie title"]})
all_rank.index += 1
all_rank = all_rank.rename_axis("Rank")
all_rank
```
